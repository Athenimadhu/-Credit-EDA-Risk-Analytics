```python
# Univariate analysis for categories

# Count plotting in logarithmic scale

def uniplot(apd,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(apd[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = apd, x= col, order=apd[col].value_counts().index,hue = hue,palette='magma') 
        
    plt.show()
```

```python
# PLotting for income range

uniplot(target0_apd,col='AMT_INCOME_RANGE',title='Distribution of income range',hue='CODE_GENDER')
```
